# Task 1: Calculate the Win Rate

In this task, you need to calculate the win rate of each Pokemon using the battle history. You first import *pokemon.csv* and *battles.csv* as pandas data frames. Then, you create a new column called ‘Win Rate’ in the data frame created from pokemon.csv. The win rate of Pokemon i is defined as follows:

```python
%%latex
$$\text{WinRate}_i = \frac{\# \text{battles that pokemon i wins}}{\# \text{battles that Pokemon i is involved}}$$
```

```python
# import both csvs
pkdf = pd.read_csv('./pokemon.csv', sep=',', index_col="#")
btdf = pd.read_csv('./battles.csv', sep=',')
```

```python
pkdf
```

```python
# here we confirm no two same pokemon has battled, so total battle count can simply add two
print ('Is there have no same pokemon battled together?', btdf[ btdf['First_pokemon'] == btdf['Second_pokemon'] ].shape[0] == 0)
```

```python
# define count function
def winCount():
    return btdf['Winner'].value_counts().sort_index()

def totalBattleCount():
    return btdf['First_pokemon'].value_counts().sort_index() + btdf['Second_pokemon'].value_counts().sort_index()
```

```python
# define win rate function
def addWinRate():
    win = winCount()
    total = totalBattleCount()
    return win.div(total)
```

```python
pkdf['WinRate'] = addWinRate()
```
